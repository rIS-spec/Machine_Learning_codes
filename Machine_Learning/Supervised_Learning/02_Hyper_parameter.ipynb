```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset('iris')
```

```python
df.head()
```

```python
df['species'].unique()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop('species', axis=1)
y = df['species']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
model_KNN = KNeighborsClassifier(n_neighbors=15)
model_KNN.fit(X_train,y_train)
```

```python
model_KNN.score(X_test,y_test)
```

```python
from sklearn.svm import SVC
```

```python
model_SVC = SVC(gamma="auto")
model_SVC.fit(X_train,y_train)
```

```python
model_SVC.score(X_test,y_test)
```

## Now lets use the Grid Search CV

```python
from sklearn.model_selection import GridSearchCV
```

```python
classifier = GridSearchCV((model_SVC),{
    'C': [1,10,20,30,40],
    'kernel': ['rbf','linear'],
},cv=5,return_train_score=False)
```

```python
classifier.fit(X,y)
```

```python
classifier.cv_results_
```

```python
result = pd.DataFrame(classifier.cv_results_)
```

```python
result
```

```python
result[['param_C','param_kernel','mean_test_score']]
```

## Now lets use the Random Search CV

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
classifier_r = RandomizedSearchCV((model_SVC),{
    'C': [1,10,20,30,40],
    'kernel': ['rbf','linear'],
},n_iter=4,cv=5,return_train_score=False)
```

```python
classifier_r.fit(X,y)
```

```python
result = pd.DataFrame(classifier_r.cv_results_)
```

```python
result
```

```python
result[['param_C','param_kernel','mean_test_score']]
```
